```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
base_folder = "../metrics/"

# size = 300
size = 400

prefixes = ["EFFB7_efficientnetb7_","R18_resnet18_","R152_resnet152_","VGG11_vgg11_"]

# file_path = os.path.join(base_folder,'EFFB7_efficientnetb7_{size}_{size}_0.csv')
# file_path = os.path.join(base_folder,f'R18_resnet18_{size}_{size}_0.csv')
# file_path = os.path.join(base_folder,f'R152_resnet152_{size}_{size}_0.csv')
file_path = os.path.join(base_folder,f'VGG11_vgg11_{size}_{size}_0.csv')


# model_name = "Efficientnet"
# model_name = "Resnet18"
# model_name = "Resnet152"
model_name = "VGG11"
path_acc = f'../figures/{model_name.lower()}-{size}-acc.png'
path_loss = f'../figures/{model_name.lower()}-{size}-loss.png'

df = pd.read_csv(file_path)
df.head()
```

```python
plt.plot(df['training_acc'])
plt.plot(df['valid_acc'])
plt.title(f'{model_name} {size}x{size} - Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.savefig(path_acc)
plt.show()
```

```python
# summarize history for loss
plt.plot(df['training_loss'])
plt.plot(df['valid_loss'])
plt.title(f'{model_name} {size}x{size} - Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.savefig(path_loss)
plt.show()
```

```python
# Plot accuracy 
ROWS = 2
COLUMNS = 4
fig, ax = plt.subplots(ROWS, COLUMNS)
# fig.set_size_inches(18.5, 10.5)
fig.set_size_inches(18, 10)


sizes = [300,400]
prefixes = ["EFFB7_efficientnetb7","R18_resnet18","R152_resnet152","VGG11_vgg11"]

for i,s in enumerate(sizes):
    for j,p in enumerate(prefixes):    
        file_name = f'{p}_{s}_{s}_0.csv'
        file_path = os.path.join(base_folder,file_name)

        path_acc = f'../figures/{p.lower()}-{s}-acc.png'
        path_loss = f'../figures/{p.lower()}-{s}-loss.png'

        df = pd.read_csv(file_path)
                
        ax[i][j].plot(df['training_acc'])
        ax[i][j].plot(df['valid_acc'])
        ax[i][j].set_title(f'{p.split("_")[1]} {s}x{s} - Model Accuracy')
        ax[i][j].set_ylabel('accuracy')
        ax[i][j].set_xlabel('epoch')        
        ax[i][j].legend(['Train', 'Validation'], loc='upper left')

# fig.savefig("../figures/accuracies.png") 
fig.savefig("../figures/accuracies.jpg")        
```

```python
# Plot loss 
ROWS = 2
COLUMNS = 4
fig, ax = plt.subplots(ROWS, COLUMNS)
fig.set_size_inches(18, 10)


sizes = [300,400]
prefixes = ["EFFB7_efficientnetb7","R18_resnet18","R152_resnet152","VGG11_vgg11"]

for i,s in enumerate(sizes):
    for j,p in enumerate(prefixes):    
        file_name = f'{p}_{s}_{s}_0.csv'
        file_path = os.path.join(base_folder,file_name)

        path_acc = f'../figures/{p.lower()}-{s}-acc.png'
        path_loss = f'../figures/{p.lower()}-{s}-loss.png'

        df = pd.read_csv(file_path)
                
        ax[i][j].plot(df['training_loss'])
        ax[i][j].plot(df['valid_loss'])
        ax[i][j].set_title(f'{p.split("_")[1]} {s}x{s} - Model Loss')
        ax[i][j].set_ylabel('loss')
        ax[i][j].set_xlabel('epoch')        
        ax[i][j].legend(['Train', 'Validation'], loc='upper left')

fig.savefig("../figures/losses.jpg")
```
